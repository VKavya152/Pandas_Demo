```python
df=pd.read_csv('Universities.csv')
print(df)
```

```python
plt.hist(df['SAT'])

plt.xlabel('SAT')
plt.ylabel('Freq')
plt.title('Histogram')
plt.show()
```

```python
import seaborn as sns 
sns.histplot(df['Expenses'])
plt.show()
```

```python
plt.plot(df['SAT'],df['Accept'],'r+')
plt.xlabel('SAT')
plt.ylabel('Accept')
plt.title('Scatterplot')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df.iloc[:,1:4])
```

```python
plt.boxplot(df['SAT'], vert=False)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(df['SAT'])
```
